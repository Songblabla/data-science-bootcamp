```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
total_sales = df['Sales'][df['Order Date'].dt.year == 2017].sum()
average_sales = df['Sales'][df['Order Date'].dt.year == 2017].mean()
sd_sales = df['Sales'][df['Order Date'].dt.year == 2017].std()

print(total_sales)
print(average_sales)
print(sd_sales)
```

```python
# TODO 06 - which Segment has the highest profit in 2018

highest_profit_segment = df[ df['Order Date'].dt.year == 2018 ].groupby('Segment')['Profit'].sum()
highest_profit_segment.head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
least_sales_segment = df[ (df['Order Date'] > '2019-04-15') & (df['Order Date'] < '2019-12-31' )].groupby('State')['Sales'].sum().reset_index().sort_values('Sales').head()
least_sales_segment
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
sales_total = df['Sales'].sum()
sales_region = df[(df['Order Date'].dt.year == 2019)].query(" Region == 'West' | Region == 'Central' ")
sales_central_west = sales_region['Sales'].sum()
proportion =  sales_central_west / sales_total
print(proportion * 100)
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df_year = df[ (df['Order Date'] > '2018-12-31') & (df['Order Date'] < '2021-01-01' )]
top10_product_order = df_year['Product Name'].value_counts().head(10)
top10_sales_order = df_year.groupby('Product Name')['Sales'].agg(['sum']).sort_values('sum',ascending = False).head(10)
top10_product_order
```

```python
top10_sales_order.columns = {'Total Sales of this product'}
top10_sales_order
```
